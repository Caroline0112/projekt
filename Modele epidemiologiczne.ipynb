```python
import numpy 
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import argparse


def odes(x,t,beta,gamma,delta): 
    S,I = x
    N=S+I
    dSdt = -beta*S*I*(1/N)
    dIdt = beta*S*I*(1/N)
    return [dSdt,dIdt]

def plotdata_si(s_start,i_start, beta,gamma,delta,time,name):  
    time = int(time)
    t = numpy.linspace(0, time, time)  
    x_start=[s_start,i_start]
    sol = odeint(odes, x_start, t, args=(beta,gamma,delta))
    S,I = sol.T
    plt.title("Susceptible, infected and recovered population") 
    plt.xlabel("time")
    plt.ylabel("Simulated numbers")
    plt.yscale('linear')
    plt.plot(t, S,label='S')
    plt.plot(t, I,label='I')
    plt.legend()
    plt.show()
plotdata_si(10,3,1.34,0,0,160, "basic") 
```
